### A translator based on matrix operation

#### 1. Read the data and build the dictionary and embedding

```python
%matplotlib inline
%load_ext autoreload
%autoreload
import unicodecsv as csv
import codecs
import numpy as np
import torch
from torch import nn

max_sentence_length = 20

# indexing the words
source_sentences = []
w2i_s = {'':0}
i2w_s = {0:''}

target_sentences = []                
w2i_t = {'':0}
i2w_t = {0:''}

n1 = 1
n2 = 1
with codecs.open('data/chat.txt','r') as f:
    data = csv.reader(f,delimiter='\t')
    for row in data:
        s1 = row[0].split(' ')[:max_sentence_length]
        s2 = row[1].split(' ')[:max_sentence_length]
        source_sentences.append(s1)
        target_sentences.append(s2)

        for w in s1:
            if not w in w2i_s: 
                w2i_s[w] = n1
                i2w_s[n1] = w
                n1 += 1
        for w in s2:
            if not w in w2i_t: 
                w2i_t[w] = n2
                i2w_t[n2] = w
                n2 += 1

n_words_source = len(w2i_s)
n_words_target = len(w2i_t)
n_words = n_words_source + n_words_target
print(len(target_sentences), n_words)
```

```python
# word embeddings
word_embedding_dim = 32
sentence_embedding_dim = word_embedding_dim * max_sentence_length
word_embeddings_source = nn.Embedding(n_words_source, word_embedding_dim, max_norm=1)
word_embeddings_target = nn.Embedding(n_words_target, word_embedding_dim, max_norm=1)
```

```python
class Translator(nn.Module):
    def __init__(self, input_size, output_size, hidden_size, n_hidden_layers=3):
        super(Translator, self).__init__()
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size = hidden_size
        self.n_hidden_layers = n_hidden_layers
        self.input = nn.Linear(input_size,hidden_size)
        self.hiddens = []
        for n in range(self.n_hidden_layers):
            self.hiddens.append(nn.Linear(hidden_size, hidden_size))
        self.output = nn.Linear(hidden_size, output_size)
    def forward(self,x, detach_emb=True):
        if detach_emb:
            x = x.detach()
        x = self.input(x)
        for i in range(self.n_hidden_layers):
            x = torch.relu(self.hiddens[i](x))
        return self.output(x)

def get_sentence_embedding(S,mode='source'):
    if mode == 'source':
        idx = [w2i_s[w] for w in S]
        for i in range(len(S), max_sentence_length):
            idx.append(0)
        idx = torch.LongTensor([idx])
        return word_embeddings_source(idx).view(1,-1)
    elif mode == 'target':
        idx = [w2i_t[w] for w in S]
        for i in range(len(S), max_sentence_length):
            idx.append(0)
        idx = torch.LongTensor([idx])
        return word_embeddings_target(idx).view(1,-1)
    else:
        print('mode error.')
        return None
```

```python
# build the model
translator = Translator(input_size=sentence_embedding_dim, \
                        output_size=sentence_embedding_dim, \
                        hidden_size=512, n_hidden_layers=1)
```

```python
# train the model
emb_learning_rate = 1e-3
trans_learning_rate = 1e-3
optim_betas = (0.9, 0.999)
n_epochs = 20000
batch_size = 5
disp_interval = 500

torch.manual_seed(1)
np.random.seed(1)

criterion = torch.nn.SmoothL1Loss()

def lr_lambda(epoch):
    return 0.99**(epoch/40.0)

emb_optimizer = torch.optim.Adam(word_embeddings_source.parameters(), lr=emb_learning_rate, betas=optim_betas)
trans_optimizer = torch.optim.Adam(translator.parameters(), lr=trans_learning_rate, betas=optim_betas)
trans_scheduler = torch.optim.lr_scheduler.LambdaLR(trans_optimizer, lr_lambda, last_epoch=-1)
emb_scheduler = torch.optim.lr_scheduler.LambdaLR(emb_optimizer, lr_lambda, last_epoch=-1)

Loss = []
for epoch in range(n_epochs):
    translator.zero_grad()
    word_embeddings_source.zero_grad()
    indices = np.random.randint(0, len(source_sentences), batch_size)
    x = get_sentence_embedding(source_sentences[indices[0]], mode='source')
    y = get_sentence_embedding(target_sentences[indices[0]], mode='target')
    for i in range(1, batch_size):
        x = torch.cat((x, get_sentence_embedding(source_sentences[indices[i]], mode='source')))
        y = torch.cat((y, get_sentence_embedding(target_sentences[indices[i]], mode='target')))
    y = y.detach()
    translated = translator(x,detach_emb=True)
    loss=criterion(translated,y)
    loss.backward()
    trans_optimizer.step()
    trans_scheduler.step()
    Loss.append(loss.tolist())
    if epoch % disp_interval == 0:
        print('step {}, loss {}'.format(epoch, np.mean(Loss[-disp_interval:])))
    
```

```python
# translate test
def trans(s):
    if isinstance(s,unicode) or isinstance(s,str):
        s = s.split(u' ')[:max_sentence_length]
    x = get_sentence_embedding(s, mode='source')
    y = translator(x,detach_emb=True).view(max_sentence_length,word_embedding_dim)
    translated = []
    for i in range(max_sentence_length):
        loss = torch.mean((word_embeddings_target.weight.detach()-y[i].detach())**2,1)
        translated.append(i2w_t[torch.argmin(loss).tolist()])
    return u' '.join(translated)
```

```python
Test_source = source_sentences
Test_target = target_sentences

for i in range(200):
    print '_______________'
    print 'Player 1: '+ ' '.join(Test_source[i][:max_sentence_length])
    print 'Predicted: '+trans(Test_source[i][:max_sentence_length])
    print '     True: '+ ' '.join(Test_target[i][:max_sentence_length])
```
